```python
import customtkinter as ctk

# Set appearance and theme
ctk.set_appearance_mode("dark")  # Options: "light", "dark", "system"
ctk.set_default_color_theme("blue")  # Options: "blue", "green", "dark-blue"

# Create the main window
app = ctk.CTk()
app.title("CustomTkinter Calculator")
app.geometry("320x450")

expression = ""

def press(num):
    global expression
    expression += str(num)
    entry_var.set(expression)

def equalpress():
    global expression
    try:
        total = str(eval(expression))
        entry_var.set(total)
        expression = total  # Store result for further calculation
    except:
        entry_var.set("Error")
        expression = ""

def clear():
    global expression
    expression = ""
    entry_var.set("")

# Entry widget
entry_var = ctk.StringVar()
entry = ctk.CTkEntry(app, textvariable=entry_var, font=("Arial", 24), justify="right", width=280, height=50)
entry.pack(pady=20)

# Button layout
buttons = [
    ("7", "8", "9", "/"),
    ("4", "5", "6", "*"),
    ("1", "2", "3", "-"),
    ("0", ".", "=", "+"),
]

# Create buttons dynamically
for row in buttons:
    frame = ctk.CTkFrame(app)
    frame.pack(pady=5)
    for char in row:
        if char == "=":
            btn = ctk.CTkButton(frame, text=char, width=60, height=60, command=equalpress)
        else:
            btn = ctk.CTkButton(frame, text=char, width=60, height=60,
                                command=lambda c=char: press(c))
        btn.pack(side="left", padx=5)

# Clear button
clear_btn = ctk.CTkButton(app, text="Clear", width=280, height=50, command=clear, fg_color="red")
clear_btn.pack(pady=10)

# Run the app
app.mainloop()
```
